```python
#calculate the mean of a sample
from numpy.random import seed
from numpy.random import randn
from numpy import mean
#seed the random number generator
seed(1)
data = 5 * randn(10000) + 100
#calculate mean
result = mean(data)
print(result)
```

```python
#Box and Whisker Plot
#it can be created by pyplot.box(x)

from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot
#seed the random number generator
seed(1)
#random number drawn from Gaussian distribution
x = [randn(1000), 5*randn(1000), 10*randn(1000)]
#create box and whisker plot
pyplot.boxplot(x)
#show line plot
pyplot.show()
```

```python
#scatter plot by pyplot.scatter(x,y)

#example of scatter plot
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot

#seed the random number generator
seed(1)
#first variable
x = 20 * randn(1000) + 100

#second variable
y = x + (10 * randn(1000) + 50)

#create scatter plot
pyplot.scatter(x,y)

#show line line
pyplot.show()
```

```python
#Random Numbers
#seed is the parameter to start off the sequence.  Given the same seed
#it will produce the same sequence of numbers everytime.  the choice
#of seed does not matter.

#seed the pseduorandom number generator
from random import seed
from random import random
#seed random number generator
seed(1)
#generate some random numbers
print(random(), random(), random())

#reset the seed
seed(1)
print(random(), random(), random())
```

```python
#Random floating point value generates values in the range between 0 and 1
#specifically in the interval (0,1]. Values are from Uniform distribution

#generate random floating point values
from random import seed
from random import random
#seed the random number generator
seed(1)
#generate random numbers between 0-1

for _ in range(10):
    value = random()
    print(value)
```

```python
#Random Integer Values generate integer values in the interval [start, end]

#generate random integer values
from random import seed
from random import randint
#seed random number generator
seed(1)
#generate some integers
for i in range(10):
    value = randint(0,10)
    print (value)
```

```python
#Random Gaussian Values generates random values from Gaussian dist. with 
#Mean 0 and std dev. 1.0

#generate random gaussian values
from random import seed
from random import gauss
#seed random number generator
seed(1)

#generate some Gaussian values
for i in range(10):
    values = gauss(0,10)
    print(values)
```

```python
#Randomly choosing from a list used to randomly choose an item from list

#choose a random element from a list
from random import seed
from random import choice

#seed random number generator
seed(1)

#prepare a sequence
sequence = [i for i in range(20)]
print(sequence)

#make choices from the sequence
for i in range (5):
    selection = choice(sequence)
    print(selection)


```

```python
#Random subsample from a list(sampling without replacement) by sample()function

from random import seed
from random import sample

#seed random number generator
seed(1)

#prepare a sequence
sequence = [i for i in range (20)]
print(sequence)

#select a subset/sample without replacement
subset=sample(sequence, 5)
print(subset)




```

```python
#Randomly Shuffle a List like shuffling a deck of cards

#randomly shuffle a sequence
from random import seed
from random import shuffle

#seed random number generator
seed(1)

#prepare a sequence
sequence = [i for i in range (20)]
print(sequence)

#randomly suffle a list
shuffle(sequence)
print(sequence)

```
